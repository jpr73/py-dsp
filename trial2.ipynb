## Interactive plots

To make grpahs zoomable, etc. [see doc](https://matplotlib.org/stable/users/explain/figure/interactive.html).

Based on [this site](https://discourse.jupyter.org/t/ipython-not-recognized-when-plotting-with-matplotlib-notebook/27268) the `ipympl` was installed via `$ pip install ipympl` and used as below.  This worked on Fedora 41 with Jupyter Notebook version `7.4.0`.

To learn matplotlib see [quick start guide](https://matplotlib.org/stable/users/explain/quick_start.html#quick-start).

```python
import numpy as np
import matplotlib.pyplot as plt

fs=1e3
num_of_samples=128

# Define time array
samples=np.arange(num_of_samples)
t=(1/fs)*samples

# Define complex signal1: sin 50Hz, amplitude=5, imag part is zero
sig1=np.zeros(num_of_samples, dtype=complex)
sig1.real=5*np.sin(2*np.pi*50*t)
```

```python
%matplotlib ipympl
```

```python
fig1,ax1 = plt.subplots()
ax1.plot(sig1.real)
```

## Numpy - iterate through array

Iteration via loop is useful for functions, [see the doc here](https://numpy.org/doc/stable/user/basics.indexing.html).

```python
# Iterate through t, make a quadrate of it as an example
for ii in samples:
    t[ii]=t[ii]*t[ii]
```

```python
fig2,ax2 = plt.subplots()
ax2.plot(t)
```
