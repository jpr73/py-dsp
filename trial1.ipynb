```python
fs=1e3
```

```python
num_of_samples=128
```

```python
# Define time array
samples=np.arange(num_of_samples)
t=(1/fs)*samples
```

```python
# Print for debug
#print(samples)
#print(t)
```

```python
# Define complex signal1: sin 50Hz, amplitude=5, imag part is zero
sig1=np.zeros(num_of_samples, dtype=complex)
sig1.real=5*np.sin(2*np.pi*50*t)

# Signal definition not by absolute freq. but by number of periods (5x now) for entire number of samples.
#sig1.real=5*np.sin(2*np.pi*((5*fs)/num_of_samples)*t)

```
